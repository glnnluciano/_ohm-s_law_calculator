Exercise 2: Ohmâ€™s Law Calculator (Object Oriented Programming Version)

```python
class Ohms_Law_Calculator:
    def __init__(self) -> None:
        self.option = ""

    def get_option(self):
        while True:
            self.option = input("What do you want to calculate? Enter 'V' if Voltage, 'I' if Current, or 'R' if Resistance.\n").lower()
            if self.option in ["v", "i", "r"]:
                break
            else:
                print(f"\n \033[91m Invalid Input: Please choose from the options. \033[0m \n")
    
    def calculate_voltage(self):
        while True:
            try:
                current = float(input("Enter the Current: "))
                if current <= 0:
                    print(f"\n \033[91m Current must be positive. \033[0m \n")
                    continue

                resistance = float(input("Enter the Resistance: "))
                if resistance <= 0:
                    print(f"\n \033[91m Resistance must be positive. \033[0m \n")
                    continue
            
                voltage = current * resistance
                print(f" \033[94m The voltage is {voltage: .2f} volts. \033[0m")
                break

            except ValueError:
                print(f"\n \033[91m Invalid Input: Please Try Again. \033[0m \n")

    def calculate_current(self):
        while True:
            try:
                voltage = float(input("Enter the Voltage: "))
                if voltage <= 0:
                    print(f"\n \033[91m Voltage must be positive. \033[0m \n")
                    continue

                resistance = float(input("Enter the Resistance: "))
                if resistance <= 0:
                    print(f"\n \033[91m Resistance must be positive. \033[0m \n")
                    continue

                current = voltage / resistance
                print(f"\033[94m The current is {current: .2f} amperes. \033[0m")
                break

            except ValueError:
                print(f"\n \033[91m Invalid Input: Please Try Again. \033[0m \n")
        
    def calculate_resistance(self):
        while True:
            try:
                voltage = float(input("Enter the Voltage: "))
                if voltage <= 0:
                    print(f"\n \033[91m Voltage must be positive. \033[0m \n")
                    continue

                current = float(input("Enter the Current: "))
                if current <= 0:
                    print(f"\n \033[91m Current must be positive. \033[0m \n")
                    continue
                
                resistance = (voltage / current)
                print(f"\033[94m The resistance is {resistance: .2f} ohms. \033[0m")
                break
            
            except ValueError:
                print(f"\n \033[91m Invalid Input: Please Try Again. \033[0m \n")

    def run(self):
        self.get_option()

        if self.option == "v":
            self.calculate_voltage()
        elif self.option == "i":
            self.calculate_current()
        elif self.option == "r":
            self.calculate_resistance()


calculator = Ohms_Law_Calculator()
calculator.run()
```
