```python
    
while True:
    option = input("What do you want to calculate? Enter 'V' if Voltage, 'I' if Current, or 'R' if Resistance.\n")

    option = option.lower()
    options = ["i", "v", "r"]

    if option not in options:
        print(f"\n \033[91m Invalid Input: Please choose from the options. \033[0m \n")
    else: 
        break

if option == 'v':
    while True:
        try:
            current = float(input("Enter the Current: "))
            if current <= 0: 
                print(f"\n \033[91m Current must be positive. \033[0m \n")
                continue

            resistance = float(input("Enter the Resistance: "))
            if resistance <= 0:
                print(f"\n \033[91m Resistance must be positive. \033[0m \n")
                continue  

            voltage = current * resistance
            print(f" \033[94m The voltage is {voltage: .2f} volts \033[0m")
            break  

        except ValueError:
            print(f"\n \033[91m Invalid Input: Please Try Again \033[0m \n")

elif option == 'i':
    while True:
        try:
            voltage = float(input("Enter the Voltage: "))
            if voltage <= 0:
                print(f"\n \033[91m Voltage must be positive. \033[0m \n")
                continue

            resistance = float(input("Enter the resistance: "))
            if resistance <= 0:
                print(f"\n \033[91m Resistance must be positive. \033[0m \n")
                continue
            
            current = voltage / resistance
            print(f"\033[94m The current is {current: .2f} amperes \033[0m")
            break

        except ValueError:
            print(f"\n \033[91m Invalid Input: Please Try Again \033[0m \n")
                
elif option == "r":
    while True:
        try:
            voltage = float(input("Enter the Voltage: "))
            if voltage <= 0:
                print(f"\n \033[91m Voltage must be positive. \033[0m \n")
                continue

            current = float(input("Enter the Current: "))
            if current <= 0:
                print(f"\n \033[91m Current must be positive. \033[0m \n")
                continue
            
            resistance = voltage / current
            print(f"\033[94m The resistance is {resistance: .2f} ohms \033[0m")
            break
        
        except ValueError:
            print(f"\n \033[91m Invalid Input: Please Try Again \033[0m \n")
```
